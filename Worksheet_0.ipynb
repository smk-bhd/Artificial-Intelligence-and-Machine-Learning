TASK - 1

Create a Python program that converts between different units of measurement.

```python
def length_conversion(value, unitfrom, unitto):
    if unitfrom == 'm' and unitto == 'ft':
       return value * 3.28084
    elif unitfrom == 'ft' and unitto == 'm':
       return value / 3.28084
    else:
       return None

def weight_conversion(value, unitfrom, unitto):
    if unitfrom == 'kg' and unitto == 'lb':
       return value * 2.20462
    elif unitfrom == 'lb' and unitto == 'kg':
       return value / 2.20462
    else:
       return None

def volume_conversion(value, unitfrom, unitto):
    if unitfrom == 'l' and unitto == 'gal':
       return value * 0.264172
    elif unitfrom == 'gal' and unitto == 'l':
       return value / 0.264172
    else:
       return None

def main():

    print("Select the type of conversion: ")
    print("1. Length (meters to feet)")
    print("2. Weight (kilograms to pounds)")
    print("3. Volume (liters to gallons)")

    try:
        choice = int(input("Enter your choice (1,2,3): "))
        if choice not in [1, 2, 3]:
            print("Invalid!Please select a valid option.")
            return

        value = float(input("Enter the value you want to convert: "))

        if choice == 1:
            print("1. Convert meters to feet")
            print("2. Convert feet to meters")
            length_choice = int(input("Enter your choice (1/2): "))
            unitfrom = "m" if length_choice == 1 else "ft"
            unitto = "ft" if length_choice == 1 else "m"
            result = length_conversion(value, unitfrom, unitto)

        elif choice == 2:
            print("1. Convert kilograms to pounds")
            print("2. Convert pounds to kilograms")
            weight_choice = int(input("Enter your choice (1/2): "))
            unitfrom = "kg" if weight_choice == 1 else "lb"
            unitto = "lb" if weight_choice == 1 else "kg"
            result = weight_conversion(value, unitfrom, unitto)

        elif choice == 3:
            print("1. Convert liters to gallons")
            print("2. Convert gallons to liters")
            volume_choice = int(input("Enter your choice (1/2): "))
            unitfrom = "l" if volume_choice == 1 else "gal"
            unitto = "gal" if volume_choice == 1 else "l"
            result = volume_conversion(value, unitfrom, unitto)

        print(f"Converted Value: {result:.4f}")

    except ValueError:
        print("Invalid input! Please enter a numeric value.")


if __name__ == "__main__":
    main()

```

TASK - 2

Create a Python program that performs various mathematical operations on a list of numbers.

```python
def sum(numbers):
    """
    Calculate the sum of a list of numbers using a for loop.

    Parameters:
    numbers (list): A list of numeric values.

    Returns:
    float: Sum of numbers.
    """
    total = 0
    for num in numbers:
        total += num
    return total

def average(numbers):
    """
    Calculate the average of a list of numbers using a for loop.

    Parameters:
    numbers (list): A list of numeric values.

    Returns:
    float: Average value.
    """
    total = 0
    count = 0
    for num in numbers:
        total += num
        count += 1
    return total / count if count > 0 else 0

def maximum(numbers):
    """
    Find the maximum number in a list using a for loop.

    Parameters:
    numbers (list): A list of numeric values.

    Returns:
    float: Maximum number.
    """
    max_value = numbers[0]
    for num in numbers:
        if num > max_value:
            max_value = num
    return max_value

def minimum(numbers):
    """
    Find the minimum number in a list using a for loop.

    Parameters:
    numbers (list): A list of numeric values.

    Returns:
    float: Minimum number.
    """
    min_value = numbers[0]
    for num in numbers:
        if num < min_value:
            min_value = num
    return min_value

def main():
    """
    Main function to prompt user input and perform calculations.
    """
    print("Math Operations on a List of Numbers")
    print("Choose an operation:")
    print("1. Sum")
    print("2. Average")
    print("3. Maximum")
    print("4. Minimum")

    try:
        choice = int(input("Enter your choice (1/2/3/4): "))
        if choice not in [1, 2, 3, 4]:
            print("Invalid choice! Please select a valid option.")
            return

        numbers = input("Enter numbers separated by spaces: ").split()
        numbers = [float(num) for num in numbers]

        if not numbers:
            print("Error: No numbers provided.")
            return

        if choice == 1:
            result = sum(numbers)
            operation = "Sum"
        elif choice == 2:
            result = average(numbers)
            operation = "Average"
        elif choice == 3:
            result = maximum(numbers)
            operation = "Maximum"
        else:
            result = minimum(numbers)
            operation = "Minimum"

        print(f"{operation} of given numbers: {result:.4f}")

    except ValueError:
        print("Invalid input! Please enter numeric values only.")


if __name__ == "__main__":
    main()
```

Nested List

```python
def extract_everyother(list):
  """Extract every other element from a list."""
  new_list = []
  i=0
  while i < len(list):
    new_list.append(list[i])
    i += 2
  return new_list

def sub_list(list, start, end):
  """ Returns a sublist from given list"""
  new_list = []
  i=start
  while i <= end:
    new_list.append(list[i])
    i += 1
  return new_list

def reverse(list):
  """Reverse a list."""
  new_list = []
  i=len(list)-1
  while i >= 0:
    new_list.append(list[i])
    i -= 1
  return new_list

def remove_firstlast_element(list):
  """Remove the first and last elements from a list."""
  new_list = []
  i=1
  while i < len(list)-1:
    new_list.append(list[i])
    i += 1
  return new_list

def get_n_elements(list,n):
  """Get the first n elements from a list."""
  new_list = []
  i=0
  while i < n and i<len(list):
    new_list.append(list[i])
    i += 1
  return new_list

def get_lastn(list,n):
  """Get the last n elements from a list."""
  new_list = []
  i=len(list)-n
  while i < len(list):
    new_list.append(list[i])
    i += 1
  return new_list

def skip_reverse(list):
  """Extracts elements in reverse order, starting from the second-to-last element and skipping one element."""
  new_list = []
  i=len(list)-2
  while i >= 0:
    new_list.append(list[i])
    i -= 2
  return new_list

def getuser_list():
  """Prompt user to enter number of elements and list elements."""
  while True:
    try:
        n=int(input("Enter number of elements in the list: "))
        if n<= 0:
          print("Enter positive integer.")
          continue
        break
    except ValueError:
        print("Invalid input! Please enter a valid integer.")

  user_list = []
  print(f"Enter {n} elements one by one:")
  for i in range(n):
    while True:
      try:
        num = int(input(f"Enter element {i + 1}: "))
        user_list.append(num)
        break
      except ValueError:
        print("Invalid input! Please enter a number.")
  return user_list

if __name__ == "__main__":
    user_list = getuser_list()

    print("\nOperations on the list:")
    print("Every other element:", extract_everyother(user_list))

    while True:
        try:
            start = int(input("\nEnter start index for sublist: "))
            end = int(input("Enter end index for sublist: "))
            if 0 <= start <= end < len(user_list):
                break
            else:
                print("Invalid range! Start should be <= end and within list size.")
        except ValueError:
            print("Invalid input! Please enter numbers.")

    print("Sublist:", sub_list(user_list, start, end))
    print("Reversed list:", reverse(user_list))
    print("Without first & last:", remove_firstlast_element(user_list))

    while True:
        try:
            n = int(input("\nEnter number of elements to extract from the start: "))
            if n > 0:
                break
            else:
                print("Please enter a positive integer.")
        except ValueError:
            print("Invalid input! Please enter a valid integer.")

    print("First", n, "elements:", get_n_elements(user_list, n))

    while True:
        try:
            n = int(input("\nEnter number of elements to extract from the end: "))
            if n > 0:
                break
            else:
                print("Please enter a positive integer.")
        except ValueError:
            print("Invalid input! Please enter a valid integer.")

    print("Last", n, "elements:", get_lastn(user_list, n))
    print("Reverse skipping:", skip_reverse(user_list))

```

Problem - 4: Matrix Operations

```python
A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

A_inverse = np.linalg.inv(A)
check = np.dot(A, A_inverse)
print("A * A⁻¹ = I:\n", check)

AB = np.dot(A, B)
BA = np.dot(B, A)
print("\nMatrix AB:\n", AB)
print("\nMatrix BA:\n", BA)
if not np.array_equal(AB, BA):
    print("\nAB is NOT equal to BA")

AB_Transpose = np.transpose(AB)
BT_AT = np.dot(B.T, A.T)
print("\n(AB)ᵀ:\n", AB_Transpose)
print("\nBᵀ * Aᵀ:\n", BT_AT)
if np.array_equal(AB_Transpose, BT_AT):
    print("\n(AB)ᵀ is equal to Bᵀ * Aᵀ")

A_Matrix = np.array([[2, -3, 1],
                     [1, -1, 2],
                     [3, 1, -1]])

B_matrix = np.array([-1, -3, 9])

A_inverse = np.linalg.inv(A_Matrix)
solution = np.dot(A_inverse, B_matrix)
print("\nLinear Equations (x, y, z):\n", solution)
```
